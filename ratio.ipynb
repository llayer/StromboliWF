```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
wf = pd.read_hdf('data/ratio_wf.h5')
wf_som = pd.read_hdf('data/ratio_wf_som.h5')
full = pd.read_hdf('data/ratio_full.h5')
full_som = pd.read_hdf('data/ratio_som_full.h5')
infra = pd.read_hdf('data/ratio_infra.h5')
infra_som = pd.read_hdf('data/ratio_som_infra.h5')
center = pd.read_hdf('data/center.h5')
```

```python
fig = plt.figure(figsize=(8, 5), dpi=100)
ax1 = plt.subplot(111)

ax1.plot(center, wf, label = 'VLP signal')
ax1.plot(center, wf_som, label = 'SOM VLP signal ')
ax1.plot(center, full, label = 'VLP signal + LPC + Amplitude')
ax1.plot(center, full_som, label = 'SOM VLP signal + LPC + Amplitude')
ax1.plot(center, 1/infra, label = 'VLP signal + LPC + Amplitude + Infra')
ax1.plot(center, 1/infra_som, label = 'SOM VLP signal + LPC + Amplitude + Infra')



start, end = 1.56009035e+18, 1.56838442e+18
ax1.set_xlim(start, end)
ax1.axhline(1, color='k', linestyle='dashed', linewidth=1)
#ax1.xaxis.set_ticks(np.arange(start, end, (end - start / 10)))
ax1.set_xticklabels([datetime.fromtimestamp(ts / 1e9).strftime('%D') for ts in ax1.get_xticks()])
ax1.set_ylabel("Ratio VLP Cls1/Cls2",  fontsize=15)
ax1.set_yscale('log')
#ax1.xaxis.set_major_locator(plt.MaxNLocator(10))
ax1.set_ylim([0.1,10])
plt.xticks(rotation=45)

plt.legend()
```
